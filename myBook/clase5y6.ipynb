# Borrado de un proceso de Poisson
Vamos a ver una operación interesante en el proceso de Poisson, que me vuelve a dar procesos de poisson.

Imganinemos que usamos un proceso de Poisson para contar eventos, por ejemplo, los coches que pasan por una carretera. A veces, me puede interesar no contabilizar todos los evento. Por ejemplo, puede querer contabilizar solo los coches con execeso de velocidad.

Por lo que al final quedaria una serie de sucesos contados y una series de sucesos sin contar. Por eso se llama el *borrado* de un proceso de Poisson, porque borro algunos sucesos.

Imaginemos que un suceso concreto lo contabilizo con probabilidad $p$, y que la contabilización de estos sucesos se realiza de modo independiente.

En esta situación, si tengo un proceso de Poisson original $(N(t), t \geq 0)$ (contar coches).

Luego tendremos $(N_1(t), t \geq 0)$ y $(N_2(t), t \geq 0)$, los procesos que describen el $N_1$, los sucesos contabilizados, y $N_2$, los no contabilizados.

Tenemos un resultado que nos dice que $ \forall t$  $N_1$ y  $N_2$ son variables independientes, y cada una de ellas tiene distribución de poisson de media $\lambda pt$ y $\lambda (1-p)t$. Es decir, los sucesos contabilizados me vuelven a dar una distribución de Poisson de tasa corregida, y los sucesos no contabilizados también.

# Ejemplo

La llegada de clientes a una tiensa se puede describir mediante un proceso de Poisson de tasa $\lambda$. Con probabilidad $p$, cada cliente compra (esta decisión es independiente de lo que hagan los demás clientes). Si en una hora han llegado 5 clientes.

1. ¿Cuál es la probabilidad de que hayan comprado exactamente 3?
2. Bajo el supuesto del enunciado, ¿qué distribución tiene la variable 'número de clientes que han comprado'?


Solución 1.

Tengo un proceso de Poisson general  $(N(t), t \geq 0)$ que se me descompone en dos:

1. $N_1(t)$ clientes que llegan y compran, $Poiss(\lambda p)$
2. $N_2(t)$ clientes que llegan y no compran, $Poiss(\lambda (1-p))$

La pregunta me dice que en una hora han llegado 5 clientes, y que de ellos han comprado 3. Por tanto sé que $N(1)=5$ y que $N_1(1)=3$. 

Probabilidad de que, dada la llegada de 5 clientes, 3 de ellos hayan comprado:
$$P(N_1(1)=3|N(1)=5)$$

Por probabilidad condicionada:
$$P(N_1(1)=3|N(1)=5) = \frac{P(N_1(1)=3,N(1)=5)}{P(N(1)=5)}$$

Como los sucesos $N_1$ y $N_2$ son independientes, si han comprado 3 clientes,2 de ellos **no** han comprado, por lo que puedo escribir:
$$P(N_1(1)=3,N(1)=5) = P(N_1(1)=3) \cdot P(N_2(1)=2)$$

Lo que da lugar a:
$$P(N_1(1)=3|N(1)=5) =  \frac{ P(N_1(1)=3) \cdot P(N_2(1)=2)}{P(N(1)=5)}$$

$$ =  \frac{\frac{e^{- \lambda \cdot p}(\lambda \cdot p )^3}{3!} \frac{e^{- \lambda \cdot (1-p)}(\lambda \cdot (1-p) )^2}{2!}}{\frac{e^{- \lambda }(\lambda  )^5}{5!}} = \binom{5}{3} p^3 (1-p)^2 $$

El resultado es el peso de una binomial, con número de ensayos 5 -los clientes que van- y con probabilidad de éxito -probabilidad de comprar- $p$.

Solución 2.

Habría que generalizar lo que hemos hecho para el caso $3$, al caso $k$:
$$P(N_1(1)=k|N(1)=5) =  \binom{5}{k} p^k (1-p)^{n-k}$$

La distribución que tiene la v.a. "número de clientes que han comprado" es una binomial de parámetros 5 y $p$, $Binom(5,p)$

# Proceso de Poisson no homogeneo

En el proceso de Poisson que hemos visto hasta ahora consideramos que la tasa de llegadas es constante a lo largo del tiempo. Esta suposición puede no ser realista. 

Por ejemplo en una tienda puede haber horas de concentración de clientes, o puede haber horas donde prácticamente no haya nadie.

Entonces deberíamos pensar en una generalización del proceso de Poisson que contemplase esa posibilidad.

Para ello vamos a modificar la definción dad anteriormente sobre el proceso de Poisson. Deciámos que el proceso de Poisson $N(t), t \leq 0)$ es un proceso de Poisson de tasa $\lambda$ si verifica: 
1. $N(0) = 0$
2. El proceso tiene incrementos estacionarios e independientes
3. $P(N(h) = 1) = \lambda h + o(h)$ (la probabilidad de que en un incremento infinitesimal ocurra 1 suceso es igual a $\lambda h + o(h)$)
4. $P(N(h) \leq 2) = o(h)$ (la probabilidad de que en un incremento infinitesimal ocurran 2 sucesos es irrelevante, lo que se representa con $o(h)$)

Modificaremos estas propiedades para generalizar el proceso de Poisson:
1. Se queda igual.
2. Los incrementos son independientes, pero ya no son estacionarios.
3. La probabilidad de que en un incremento infinitesimal ocurra 1 suceso es igual a la $\lambda$ de ese incremento infinitesimal  $\lambda (t)$ más una $o$ pequeña de $h$,  $o(h)$ $\rightarrow P(N(t+h)-N(t)=1)= \lambda (t) +o(h)$
4. La probabilidad de que en un incremento infinitesimal ocurran dos sucesos es una $o$ pequeña de $h$ $\rightarrow P(N(t+h)-N(t) \leq 2)=o(h)$

Ahora ya tengo una generalización del proceso de Poisson que me permitirá tratar situaciones donde la tasa de llegadas dependa del tiempo.




Una vez tengo la generalización, ahora tengo que saber como se hacen los cálculos. Es decir, como se distribuyen los incrementos.

# El proceso de Poisson Compuesto

Otra generalización del proceso de Poisson que vamos a explicar con un ejemplo:

Imaginad que tenemos un proceso de proceso que describe las llegadas de clientes a una tienda. 
$$N(t) = \{ \text{llegadas de clientes a una tienda}\} $$

LLega el cliente y compra. Y puedo considerar la cantidad que se gasta. Es decir, para cada cliente $i$, puedo considerar que compra una cantidad de $Y_i$ €. Si sumo las compras de los clientes en un intervalo de tiempo $t$, obtendré

$$\sum_{i=1}^{N(t)}Y_i =  \{ \text{cantidad gastada por clientes hasta el instante t}\}  $$

Considero un proceso de Poisson, considero que cada llegada lleva asociada una v.a. que en este caso es lo que gasta cada cliente, y tendré un nuevo proceso que será esta suma de variables aleatorias. Esto es un proceso de Poisson compuesto. 

Hacer cálculos con un proceso de poisson compuesto es complicado porque hay que hacer cálculos sobre sumas aleatorias. Por lo que vamos a hacer un inciso para ver qué es una suma aleatoria, qué cálculos se pueden hacer con ella y cuales no. 

## Sumas aleatorias
### Funciones generatricez de momentos

El interés que tiene la función generatriz de momentos es que determina univocamente la distribución de partida. Entonces hay casos de variables aleatorias cuyos cálculos de probabilidades son muy dificiles de realizar, pero que son más fáciles de realizar los de la función generatriz, y si identifico cual es su función generatriz ya se de qué v.a. estoy hablando.

$$\psi (t) = Ee^{t X}$$

Vamos a hacer un ejemplo antes de continuar. Dijimos que la suma de exponenciales independientes es $\Gamma$:

La suma de exponenciales independientes de parámetro $\lambda$ $X_1 +...+X_n$ tiene distribución  $\Gamma$ de parámetro de forma $n$ y parametro de escala $\lambda$. Esto lo puedo demostrar calculanda la funcion generatriz.

Voy a llamar $\Psi_n$ a la función generatriz de esta suma $X_1 +...+X_n$

y voy a llama $\Psi$ a la función generatiz de cada sumando, que es común.

Entonces voy a calcular la función generatriz de la suma, la cual me interesa ver que tiene distribución  $\Gamma$.

$$\Psi_n = E[e^{t (X_1 +...+X_n)}] = E[e^{tX_1} \cdot e^{tX_2} \cdot...\cdot e^{tX_n}]$$

Al ser variables independientes, la esperanza del producto es lo mismo que el producto de la esperanza:

$$E[e^{tX_1}] \cdot E[e^{tX_2}] \cdot ... \cdot E[e^{tX_n}]  = \Psi(t)^n  $$

Es decir: la función generatriz de una suma de variables independientes es la función generatriz de las variables originales elevado a $n$, independientemente de la distribución de los sumandos.

$$\Psi_n(t)=\Psi(t)^n$$

Sé que la función generatiz de una exponencial $\psi(t) = \frac{\lambda}{\lambda-t}$. Entonces la función generatriz de la suma es $\Psi(t)^n = (\frac{\lambda}{\lambda-t})^n$




**Seguimos después del descanso**

Ya hemos dicho que la función generatriz de momentos, aparte de generar los momentos, me determina univocamente la distribución de partida.

Normalmente estamos trabajando con variables no negativas (poisson, binomial, gamma...), entonces la función generatriz de momentos existe sin problemas.

Entonces, para trabajan con la función generatriz, necesitámos la existencia de dicha función. Pero como no siempre existe, vamos a definir una función que tiene las mismas propiedades pero que siempre existe, **la función característica**.

Dada una variable aleatoria continua $X$ su función característica es:
$$\phi_{X(t)} = E[e^{itX}]$$

Ventajas:
* Pasamos a variable compleja, pero esta función siempre existe.
* Propiedades análogas a la función generatriz: Determina univocamente la distribución de partida, y calculando sus derivadas podemos calcular los momentos de la variable.

### Sumas aleatorias: distribuciones compuestas.

Son las que nos van a ayudar a calcular lo que queramos calcular en un proceso de Poisson compuesto.

LLamaremos sumas aleatorias a sumas de variables aleatorias independientes e idénticamente distribuídas donde el número de sumandos es aleatorio.

Esto es, tenemos $Y_1, Y_2,...,Y_i...$ variables aleatorias idénticamente distribuídas no negativas.

Tenemos $N$, una v.a. que toma valores en $0,1,...,n...,$ independientemente de la sucesión anterior.

Y nos interesa estudiar la variable $S$ que es la suma de las v.a. $Y_i$ hasta el instante aleatorio $N$.

$$S:= \sum_{i=0}^{N}Y_i \hspace{2em} (\sum_{i=0}^{0}Y_i = 0)$$

Algunas aplicaciones:
* Modelos de riesgo en matemática actuarial (número de clientes que reclaman en una compañia de seguros: gasto de la compañia será $S$, e $Y_i$ serán las cantidades reclamadas)
* Investigación operativa (teoría del caos) (tiempo total que se ha invertido atendiendo a clientes en una tienda, donde $Y_i$ es la cantidad de tiempo de atención de un cliente)
* Meteorología

**Distribución compuesta** se llama a la función de distribución de una suma aleatoria. Para calcularla, lo primero que necesitaré será la distribución común de los sumandos $F$, y lo segundo serán los pesos de la distribución discreta que me define el índice de sumación, $p_n := P(N=n)$.

La función de distribución compuesta viene dada por

$$G(x) = P(X=0)+ \sum_{n=1}^{\infty}P(N=n)P(Y_1+...+Y_n \leq x)$$
$$= \sum_{n=1}^{\infty} p_n F^n (x), \hspace{1em} x \geq 0 $$

El problema es que la evaluación exacta de una distribución compuesta es un problema complicado. Requiere normalmente el uso de cotas o aproximaciones de dicha función.

Para ayudarnos en esa tarea esta la función generatriz de momentos de una distribución compuesta. Mediante esta función puedo determinar la suma aleatoria, o en el peor de los casos, por lo menos podre determinar los momentos de la suma aleatoria.

Si denotamos por $\psi(\cdot)$ a la función generatriz de momentos común para $Y_i$, tenemos que la función generatriz de la distribución compuesta viene dada por:

$$\psi_{S(t)} = E \psi(t)^N = \sum_{n=0}^{\infty}\psi (t)^n P(N=n)$$

En particular, podríamos deducir, derivando la anterior expresión que:

* $E[\sum_{i=0}^N Y_i] = E[N] \cdot E[Y_i]$
* $Var[\sum_{i=0}^N Y_i] = Var[N](E[Y_1])^2 + E[N]Var[Y_i]$

## Proceso de Poisson condicional
Otra generalización interesante del proceso de Poisson es el proceso de Poisson condicional. Aquí lo que hacemos es pensar que la tasa no se mantiene fija, sino que puede ser una variable aleatoria. 

Es decir, consideramos $\Lambda$ una variable aleatoria postivia con función de distribución $G$, que representa la tasa aleatoria de este proceso. Entonces, consideramos $(N(t), t > 0)$ un un proceso de recuento de modo que cuando sabemos que esta v.a. vale $\lambda$, entonces $(N(t), t > 0)$ es un proceso de Poisson de tasa $\lambda$. 

Se le llama proceso de poisson condicional porque condicinado a $\lambda$ sabemos si es un proceso de Poisson.

Si quisiera hacer el cálculo de la probabilidad de que un incremento dado valga $n. $:

$$P(N(t+s)-N(t)=n)=E \frac{e^{ \Lambda s}(\Lambda s)^n}{n!}$$
$$= \int_0^{\infty} \frac{e^{ \lambda s}(\lambda s)^n}{n!}dG(\lambda), \hspace{1em} n=0,1,...$$


Si $\Lambda$  es discreta, esta integral es $\sum_{i=0}^k f(\lambda_k)P(\Lambda = \lambda_k)$ para $ \lambda_1, \lambda_2,...,\lambda_k$

Si $\Lambda$  es continua, esta integral es $\int_0^{\infty}f(\lambda)g(\lambda)d \lambda$

donde $f(\lambda) = \frac{e^{ \lambda s}(\lambda s)^n}{n!}$



Razones para considerar esta extensión:

Sabemos que en un proceso de Poisson la esperanza es igual a la varianza. Por tanto $\frac{Var(N(t))}{E(N(t))}=1$. Si estamos observando empiricamente un proceso y vemos que esta relación es mayor que 1, se dice que hay una sobredispersión. Por tanto una forma de buscar un proceso que verifique esta propiedad es considerar que esa tasa $\lambda$ es aleatoria y que hay mayor variabilidad que en un proceso de Poisson.

En concreto, en el caso de modelizar reclamaciones en una compañia de seguros, esa sobredispersión puede deberse a la heterogeneidad entre los grupos de asegurados (diferencias en las tasas según sexo, edad...).





## Ejemplo

Supongamos que la tasa con la que ocurren terremotos en una región dada (y periodo de tiempo concreto), es variable. Hay periodos muy largos en los que esa tasa permanece constante, pero en otros, esa tasa varía

Por razones que se desconocen, hay periodos de mayor riesgo sísmico,  $\lambda_2$,  y de menor riesgo sísmico, $\lambda_1$.

dicha tasa es de $\lambda_1$ el $100p$% de las ocasiones y $\lambda_2$ el $100(1-p)$%

Un modelo razonable para dicha situación es pensar en $(N(t), t >0)$ un proceso de Poisson condicional donde la tasa aleatoria $\Lambda$ toma los valores $\lambda_1$ con probabilidad $p$ y $\lambda_2$ con probabilidad $1-p$ 

Si han ocurrido $n$ terremotos en los primeros $t$ instantes del tiempo al inicio del periodo considerado ¿cuál es la probabilidad de que dicho periodo sea un periodo de bajo riesgo sísmico $\lambda_1$? ¿Cuál s la distribución del tiempo que transcurre desde $t$ al siguiente terremoto?

Solución 1.
La probabilidad de  $\Lambda = \lambda_1$ es $p$. Pero si sé que se han producido muchos o pocos terremotos, esta probabilidad aumentará o disminuirá en función de esta información. 

Probabilidad de que dicho periodo sea un periodo de bajo riesgo sísmico $\lambda_1$, dada la ocurrencia de $n$ terremotos:
$$P(\Lambda = \lambda_1 | N(t) = n)$$
Por Bayes: 
$$\frac{P(\Lambda = \lambda_1 | N(t) = n)P(\Lambda = \lambda_1)}{ \sum_{i=1}^2 P(N(t) = n |\Lambda = \lambda_i)P(\Lambda = \lambda_1)}$$
$$$$
$$\frac{\frac{e^{- \lambda_1 t}(\lambda_1 t)^n}{n!}p}{\frac{e^{- \lambda_1 t}(\lambda_1 t)^n}{n!}p + \frac{e^{- \lambda_2 t}(\lambda_2 t)^n}{n!}(1-p)}$$

La probabilidad de tener un periodo tranquilo me vendra determinada por el número de sucesos que hayan ocurrido hasta el instante $t$. Si son muchos, esta probabilidad será más baja y si son pocos la probabilidad será más alta.

Solución 2.

Distribución que transcurre desde un instante determinado $t$ hasta el siguiente terremoto. Para ello defino la variable $T$:

$$T = \{ \text{tiempo desde t hasta el siguiente terremoto} \}$$

En un proceso de Poisson normal sé que esta $T$ sería exponencial, pero en un proceso de Poisson condicional no es así.

Quiero calcular la probabilidad de que el instante $T$ sea mayor que $s$. 

Estoy en el instante $t$, tomo un intervalo de longitud $s$ y quier saber y quiero saber cual es la probabilidad de que el tiempo hasta el siguiente terremoto $T$ valga una cantidad mayor que $s$

Como han ocurrido $n$ terremotos en los primeros $t$ instantes de tiempo, querré calcular $P(T>s | N(t) = n)$

Esto lo puedo poner: $$\frac{P(T>s , N(t) = n)}{P(N(t)=n)}$$

Esto lo puedo poner en términos del proceso y de sus incrementos:La probabilidad de $P(T>s , N(t) = n)$ es lo mismo que en el incremento hayan pasado 0 terremotos, $(N(t+s)-N(t)=0)$:

$$\frac{P(N(t+s)-N(t)=0 , N(t) = n)}{P(N(t)=n)}$$

Aquí, **no puedo aplicar la propiedad de incrementos independientes**. Pero lo puedo calcular condicionando al velor de $\Lambda$:
$$P(N(t+s)-N(t)=0 , N(t) = n) = \sum_{i=1}^2 P((N(t+s)-N(t)=0 , N(t) = n)|\Lambda = \lambda_i)P(\Lambda = \lambda_i) $$

Cuando condiciono el valor de $\Lambda$, ya sé que $(N(t+s)-N(t)=0 , N(t) = n)$ es un proceso de Poisson, y ya puedo aplicar la propiedad de incrementos independientes.
